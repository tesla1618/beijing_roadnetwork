```python
# Define the place (Beijing, China)
place_name = "Beijing, China"

# Download the road network graph
G = ox.graph_from_place(place_name, network_type="all")

# Create a DataFrame with road segment ID, start node ID, and end node ID
df = ["Road segment ID", "Start Node ID", "End Node ID", 'data']
```

```python
for u, v, data in G.edges(data=True):
    road_segment_id = data["osmid"]
    start_node_id = u
    end_node_id = v
    df = df.append(
        {
            "Road segment ID": road_segment_id,
            "Start Node ID": start_node_id,
            "End Node ID": end_node_id,
            "data": data,
        },
        ignore_index=True,
    )
```

```python
# Get node coordinates
nodes, _ = ox.graph_to_gdfs(G, edges=False)
df["Start Latitude"] = nodes.loc[df["Start Node ID"]]["y"].values
df["Start Longitude"] = nodes.loc[df["Start Node ID"]]["x"].values
df["End Latitude"] = nodes.loc[df["End Node ID"]]["y"].values
df["End Longitude"] = nodes.loc[df["End Node ID"]]["x"].values


# Display the resulting DataFrame
print(df)
```
